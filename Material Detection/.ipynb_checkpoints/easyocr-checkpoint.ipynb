```python
from ultralytics import YOLO
import cv2
from PIL import Image
import easyocr 
# Set the path to the Tesseract executable


# Path to the image and the model
img_path =("C:\\Users\\almas\\Desktop\\opq.jpg")
img = cv2.imread(img_path)
H, W, _ = img.shape
model_path = "C:\Users\almas\Documents\Projects\\Specs\\Model\\best.pt"
model = YOLO(model_path)  

# Define the detection threshold
threshold = 0.5
reader=easyocr.Reader(['en'],gpu=True)

# Perform inference
results = model(img)[0]


# Loop through the detection results
for result in results.boxes.data.tolist():
    x1, y1, x2, y2, score, class_id = result
    if score > threshold:
        # Draw the bounding box
        cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        # Put the class name
        cv2.putText(img, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
        width = int(x2) - int(x1)
        height = int(y2) - int(y1)
        
        # Extract the region of interest (ROI) for text extraction
        roi = img[int(y1):int(y2), int(x1):int(x2)]
        # Convert the ROI to grayscale for better OCR results
 
        text=reader.readtext(roi)
        for t in text:
            print(f"Extracted text: {t[1]}")
            
img =cv2.resize(img,(900,600))
cv2.imshow('op',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
import pytesseract
from PIL import Image
import easyocr 
# Set the path to the Tesseract executable


# Path to the image and the model
img_path =("C:\\Users\\almas\\Desktop\\opq.jpg")
img = cv2.imread(img_path)
img =cv2.resize(img,(900,600))
H, W, _ = img.shape
model_path = "C:\\Users\\almas\\Downloads\\best.pt"
model = YOLO(model_path)  

# Define the detection threshold
threshold = 0.5
reader=easyocr.Reader(['en'],gpu=True)

# Perform inference
results = model(img)[0]


# Loop through the detection results
for result in results.boxes.data.tolist():
    x1, y1, x2, y2, score, class_id = result
    if score > threshold:
        # Draw the bounding box
        cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        # Put the class name
        cv2.putText(img, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
        width = int(x2) - int(x1)
        height = int(y2) - int(y1)
        
        # Extract the region of interest (ROI) for text extraction
        roi = img[int(y1):int(y2), int(x1):int(x2)]
        # Convert the ROI to grayscale for better OCR results
 
        text=reader.readtext(roi)
        for t in text:
            print(f"Extracted text: {t}")
            bbox,op,score = t
            print(bbox)
            if score>0:
                cv2.rectangle(roi,bbox[0],bbox[2],(0,255,0),2)



cv2.imshow('op',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
